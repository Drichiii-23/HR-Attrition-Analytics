```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All packages imported successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# Load the dataset
df = pd.read_csv('../data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Check it loaded
print(f"Dataset shape: {df.shape}")
print(f"First 5 rows:")
df.head()
```

```python
# Get comprehensive info about the dataset
print("=== DATASET INFORMATION ===\n")
print(f"Shape: {df.shape}")  # 1470 rows, 35 columns
print(f"\n=== COLUMN NAMES ===\n")
print(df.columns.tolist())
print(f"\n=== DATA TYPES ===\n")
print(df.dtypes)
print(f"\n=== MISSING VALUES ===\n")
print(df.isnull().sum())
print(f"\n=== BASIC STATISTICS ===\n")
df.describe()
```

```python
# Check unique values in text columns
text_columns = ['Department', 'BusinessTravel', 'Gender', 'MaritalStatus', 'OverTime']

for col in text_columns:
    print(f"\n=== {col} ===")
    print(df[col].value_counts())
    print(f"Unique values: {df[col].nunique()}")
```

```python
# Check categorical numeric columns
categorical_numeric = ['Education', 'EnvironmentSatisfaction', 'JobSatisfaction', 
                       'RelationshipSatisfaction', 'WorkLifeBalance', 'JobInvolvement',
                       'PerformanceRating', 'StockOptionLevel']

for col in categorical_numeric:
    print(f"\n=== {col} ===")
    print(df[col].value_counts().sort_index())
    print(f"Unique values: {df[col].nunique()}")
```

```python
# Create mapping dictionaries based on the values we just saw
education_map = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}

satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}

involvement_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}

# For PerformanceRating, we see only 3 and 4
performance_map = {3: 'Good', 4: 'Outstanding'}

stock_option_map = {0: 'None', 1: 'Low', 2: 'Medium', 3: 'High'}

print("✅ Mappings created successfully!")
print("\nEducation levels:", education_map)
print("Satisfaction levels:", satisfaction_map)
print("Performance ratings:", performance_map)
print("Stock option levels:", stock_option_map)
```

```python
# Create new columns with readable labels
df['Education_Level'] = df['Education'].map(education_map)
df['Environment_Satisfaction_Level'] = df['EnvironmentSatisfaction'].map(satisfaction_map)
df['Job_Satisfaction_Level'] = df['JobSatisfaction'].map(satisfaction_map)
df['Relationship_Satisfaction_Level'] = df['RelationshipSatisfaction'].map(satisfaction_map)
df['Work_Life_Balance_Level'] = df['WorkLifeBalance'].map(satisfaction_map)
df['Job_Involvement_Level'] = df['JobInvolvement'].map(involvement_map)
df['Performance_Rating_Level'] = df['PerformanceRating'].map(performance_map)
df['Stock_Option_Level'] = df['StockOptionLevel'].map(stock_option_map)

# Check the first few rows to verify
print("First 5 rows with new columns:")
df[['Education', 'Education_Level', 
    'JobSatisfaction', 'Job_Satisfaction_Level',
    'PerformanceRating', 'Performance_Rating_Level']].head()
```

```python
# Check all new columns' distributions
new_cols = ['Education_Level', 'Environment_Satisfaction_Level', 'Job_Satisfaction_Level',
            'Relationship_Satisfaction_Level', 'Work_Life_Balance_Level', 
            'Job_Involvement_Level', 'Performance_Rating_Level', 'Stock_Option_Level']

for col in new_cols:
    print(f"\n=== {col} Distribution ===")
    print(df[col].value_counts())
    print(f"Total: {df[col].count()}")
    print("-" * 30)
    
```

```python
# Check for any duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    print("\nShowing first few duplicate rows:")
    df[df.duplicated(keep=False)].head()
else:
    print("✅ No duplicates found! The dataset is clean.")
```

```python
# Create business-focused features

# 1. High Risk Categories (based on satisfaction)
df['High_Risk'] = ((df['Job_Satisfaction_Level'].isin(['Low', 'Medium'])) | 
                   (df['Work_Life_Balance_Level'].isin(['Low', 'Medium']))).astype(int)

# 2. Overtime Impact (already have OverTime_Flag)

# 3. Compensation Ratio (MonthlyIncome relative to average for their role)
# First, get average salary by JobRole
avg_salary_by_role = df.groupby('JobRole')['MonthlyIncome'].transform('mean')
df['Salary_vs_Role_Avg'] = df['MonthlyIncome'] / avg_salary_by_role

# 4. Long Tenure without Promotion
df['Years_Without_Promotion'] = df['YearsSinceLastPromotion']
df['Long_Without_Promo'] = (df['YearsSinceLastPromotion'] >= 3).astype(int)

# 5. Commute Impact (Distance from home categorized)
df['Commute_Distance'] = pd.cut(df['DistanceFromHome'],
                                 bins=[0, 2, 5, 10, 29],
                                 labels=['Near', 'Moderate', 'Far', 'Very Far'])

print("=== BUSINESS-FOCUSED FEATURES CREATED ===\n")
print("High Risk Employees (Low/Medium Satisfaction):")
print(df['High_Risk'].value_counts())
print(f"→ {df['High_Risk'].sum()} employees at high risk\n")

print("Long without Promotion (>3 years):")
print(df['Long_Without_Promo'].value_counts())
print(f"→ {df['Long_Without_Promo'].sum()} employees overdue for promotion\n")

print("Commute Distance Distribution:")
print(df['Commute_Distance'].value_counts())
```

```python
# Calculate key business metrics
total_employees = len(df)
total_attrition = df['Attrition'].value_counts()
attrition_rate = (total_attrition['Yes'] / total_employees) * 100

print("=== KEY BUSINESS METRICS ===\n")
print(f"Total Employees: {total_employees:,}")
print(f"Employees Left: {total_attrition['Yes']:,}")
print(f"Employees Stayed: {total_attrition['No']:,}")
print(f"Overall Attrition Rate: {attrition_rate:.1f}%\n")

# Cost impact (using conservative estimate of 50% of annual salary)
avg_salary = df['MonthlyIncome'].mean()
annual_salary = avg_salary * 12
cost_per_hire = annual_salary * 0.5  # 50% of annual salary
total_turnover_cost = total_attrition['Yes'] * cost_per_hire

print(f"Average Monthly Salary: ${avg_salary:,.0f}")
print(f"Average Annual Salary: ${annual_salary:,.0f}")
print(f"Estimated Cost per Hire: ${cost_per_hire:,.0f}")
print(f"Total Turnover Cost: ${total_turnover_cost:,.0f}")
print(f"→ Potential Savings if attrition reduced by 10%: ${total_turnover_cost * 0.1:,.0f}")

# Add this to your insights
print(f"\n⚠️  With {total_attrition['Yes']} employees leaving, and {total_employees} total employees...")
print(f"   Each 1% reduction in attrition saves ${total_turnover_cost * 0.01:,.0f}")
```

```python
# Create OverTime_Flag if it doesn't exist
if 'OverTime_Flag' not in df.columns:
    df['OverTime_Flag'] = df['OverTime'].map({'Yes': 1, 'No': 0})
    print("✅ OverTime_Flag created successfully!")
else:
    print("✅ OverTime_Flag already exists")

# Verify it worked
print(f"\nOverTime values:")
print(df['OverTime'].value_counts())
print(f"\nOverTime_Flag values:")
print(df['OverTime_Flag'].value_counts())
```

```python
# Department-level attrition analysis
dept_analysis = df.groupby('Department').agg({
    'Attrition': lambda x: (x == 'Yes').sum(),
    'EmployeeNumber': 'count',
    'MonthlyIncome': 'mean',
    'OverTime_Flag': 'mean',
    'Job_Satisfaction_Level': lambda x: (x == 'Low').sum()
}).round(2)

# Rename columns
dept_analysis.columns = ['Attrition_Count', 'Total_Employees', 'Avg_Salary', 'Overtime_Rate', 'Low_Satisfaction_Count']

# Calculate additional metrics
dept_analysis['Attrition_Rate'] = (dept_analysis['Attrition_Count'] / dept_analysis['Total_Employees'] * 100).round(1)

# Use cost_per_hire from previous calculations (if not defined, calculate it)
if 'cost_per_hire' not in locals():
    avg_salary = df['MonthlyIncome'].mean()
    annual_salary = avg_salary * 12
    cost_per_hire = annual_salary * 0.5

dept_analysis['Turnover_Cost'] = dept_analysis['Attrition_Count'] * cost_per_hire
dept_analysis['% of Total Turnover Cost'] = (dept_analysis['Turnover_Cost'] / (237 * cost_per_hire) * 100).round(1)

print("=== DEPARTMENT ANALYSIS ===\n")
print(dept_analysis.sort_values('Attrition_Rate', ascending=False).to_string())

# Find highest risk department
highest_risk_dept = dept_analysis.sort_values('Attrition_Rate', ascending=False).index[0]
print(f"\n⚠️  HIGHEST RISK DEPARTMENT: {highest_risk_dept}")
print(f"   Attrition Rate: {dept_analysis.loc[highest_risk_dept, 'Attrition_Rate']}%")
print(f"   Annual Cost: ${dept_analysis.loc[highest_risk_dept, 'Turnover_Cost']:,.0f}")
print(f"   % of Total Turnover Cost: {dept_analysis.loc[highest_risk_dept, '% of Total Turnover Cost']}%")
```

```python
# See all column names
print("Available columns:")
for i, col in enumerate(sorted(df.columns), 1):
    print(f"{i:3}. {col}")
```

```python
# Job role analysis - this is where the real insights are!
role_analysis = df.groupby('JobRole').agg({
    'Attrition': lambda x: (x == 'Yes').sum(),
    'EmployeeNumber': 'count',
    'MonthlyIncome': 'mean',
    'OverTime_Flag': 'mean',
    'Job_Satisfaction_Level': lambda x: (x == 'Low').sum(),
    'YearsAtCompany': 'mean',
    'High_Risk': 'sum'
}).round(2)

role_analysis.columns = ['Attrition_Count', 'Total_Employees', 'Avg_Salary', 'Overtime_Rate', 
                         'Low_Satisfaction', 'Avg_Tenure', 'High_Risk_Count']
role_analysis['Attrition_Rate'] = (role_analysis['Attrition_Count'] / role_analysis['Total_Employees'] * 100).round(1)
role_analysis['Turnover_Cost'] = role_analysis['Attrition_Count'] * cost_per_hire
role_analysis['High_Risk_%'] = (role_analysis['High_Risk_Count'] / role_analysis['Total_Employees'] * 100).round(1)

print("=== JOB ROLE ANALYSIS (Sorted by Attrition Rate) ===\n")
print(role_analysis.sort_values('Attrition_Rate', ascending=False).to_string())

# Identify top 3 roles to focus on
print(f"\n🎯 TOP 3 HIGHEST IMPACT ROLES FOR INTERVENTION:")
top_3_roles = role_analysis.sort_values('Attrition_Rate', ascending=False).head(3)
for i, (role, row) in enumerate(top_3_roles.iterrows(), 1):
    print(f"\n{i}. {role}")
    print(f"   • Attrition Rate: {row['Attrition_Rate']}% ({row['Attrition_Count']} employees)")
    print(f"   • Annual Cost: ${row['Turnover_Cost']:,.0f}")
    print(f"   • Overtime Rate: {row['Overtime_Rate']*100:.0f}%")
    print(f"   • High Risk Employees: {row['High_Risk_%']}% of role")
```

```python
# Compare leavers vs stayers
attrition_drivers = df.groupby('Attrition').agg({
    'MonthlyIncome': 'mean',
    'OverTime_Flag': 'mean',
    'Job_Satisfaction_Level': lambda x: (x == 'Low').sum() / len(x) * 100,
    'Work_Life_Balance_Level': lambda x: (x == 'Low').sum() / len(x) * 100,
    'YearsAtCompany': 'mean',
    'YearsSinceLastPromotion': 'mean',
    'DistanceFromHome': 'mean',
    'High_Risk': 'mean'
}).round(2)

attrition_drivers.columns = ['Avg_Salary', 'Overtime_Rate', 'Low_Satisfaction_%', 
                             'Low_WLB_%', 'Avg_Tenure', 'Years_Since_Promo', 'Avg_Commute', 'High_Risk_%']

print("=== ATTRITION DRIVERS: LEAVERS VS STAYERS ===\n")
print(attrition_drivers)

# Calculate the differences
print(f"\n=== KEY INSIGHTS - WHY EMPLOYEES LEAVE ===")
leavers = attrition_drivers.loc['Yes']
stayers = attrition_drivers.loc['No']

print(f"\n💰 COMPENSATION:")
print(f"   • Leavers earn ${stayers['Avg_Salary'] - leavers['Avg_Salary']:,.0f} LESS than stayers")
print(f"   • That's a {(1 - leavers['Avg_Salary']/stayers['Avg_Salary'])*100:.1f}% salary gap")

print(f"\n⏰ WORK-LIFE BALANCE:")
print(f"   • {leavers['Overtime_Rate']*100:.0f}% of leavers worked overtime vs {stayers['Overtime_Rate']*100:.0f}% of stayers")
print(f"   • Leavers are {(leavers['Overtime_Rate']/stayers['Overtime_Rate']-1)*100:.0f}% more likely to work overtime")

print(f"\n😞 SATISFACTION:")
print(f"   • {leavers['Low_Satisfaction_%']:.1f}% of leavers had LOW satisfaction vs {stayers['Low_Satisfaction_%']:.1f}% of stayers")
print(f"   • {leavers['Low_WLB_%']:.1f}% of leavers had LOW work-life balance vs {stayers['Low_WLB_%']:.1f}% of stayers")

print(f"\n📈 CAREER PROGRESSION:")
print(f"   • Leavers wait {leavers['Years_Since_Promo']:.1f} years for promotion vs {stayers['Years_Since_Promo']:.1f} for stayers")
print(f"   • That's {leavers['Years_Since_Promo'] - stayers['Years_Since_Promo']:.1f} years longer without promotion")

print(f"\n🚗 COMMUTE:")
print(f"   • Leavers travel {leavers['Avg_Commute']:.1f} miles vs {stayers['Avg_Commute']:.1f} miles for stayers")
```

```python
# Create all missing columns for the final dataset

# 1. Create Age_Group if it doesn't exist
if 'Age_Group' not in df.columns:
    df['Age_Group'] = pd.cut(df['Age'], 
                              bins=[18, 25, 35, 45, 55, 65], 
                              labels=['18-25', '26-35', '36-45', '46-55', '56-65'])
    print("✅ Created Age_Group")

# 2. Create Tenure_Group if it doesn't exist
if 'Tenure_Group' not in df.columns:
    df['Tenure_Group'] = pd.cut(df['YearsAtCompany'],
                                 bins=[0, 2, 5, 10, 20, 40],
                                 labels=['0-2 years', '2-5 years', '5-10 years', '10-20 years', '20+ years'])
    print("✅ Created Tenure_Group")

# 3. Create Attrition_Flag if it doesn't exist
if 'Attrition_Flag' not in df.columns:
    df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
    print("✅ Created Attrition_Flag")

# 4. Verify all columns now exist
print("\n=== CHECKING ALL REQUIRED COLUMNS ===")
required_cols = ['Age_Group', 'Tenure_Group', 'Attrition_Flag']
for col in required_cols:
    if col in df.columns:
        print(f"✅ {col} exists")
    else:
        print(f"❌ {col} MISSING")

# Show sample of new columns
print("\n=== SAMPLE OF NEW COLUMNS ===")
df[['Age', 'Age_Group', 'YearsAtCompany', 'Tenure_Group', 'Attrition', 'Attrition_Flag']].head()
```
