```python
import pandas as pd

print("=" * 50)
print("ADDING EMPLOYEENUMBER TO CLEANED DATASET")
print("=" * 50)

# Load your current Power BI dataset
print("\n📂 Loading files...")
current_df = pd.read_csv('../data/cleaned/HR_PowerBI_Ready.csv')
raw_df = pd.read_csv('../data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Check row counts
print(f"\n📊 Current dataset: {len(current_df)} rows")
print(f"📊 Raw dataset: {len(raw_df)} rows")

if len(current_df) == len(raw_df):
    print("\n✅ Row counts match!")
    
    # Check if data is in same order (compare first few rows of Age)
    print("\n🔍 Verifying data order (first 5 rows - Age):")
    match = True
    for i in range(5):
        current_age = current_df['Age'].iloc[i]
        raw_age = raw_df['Age'].iloc[i]
        print(f"   Row {i}: Current Age={current_age}, Raw Age={raw_age} - {'✓' if current_age == raw_age else '✗'}")
        if current_age != raw_age:
            match = False
    
    if match:
        print("\n✅ Data appears to be in the same order!")
        
        # Add EmployeeNumber
        current_df['EmployeeNumber'] = raw_df['EmployeeNumber']
        print("\n✅ EmployeeNumber column added!")
        
        # Show first few rows to verify
        print("\n📋 First 5 rows with new EmployeeNumber:")
        print(current_df[['Age', 'Gender', 'Department', 'EmployeeNumber']].head())
        
        # Save new file
        output_path = '../data/cleaned/HR_PowerBI_Ready_with_ID.csv'
        current_df.to_csv(output_path, index=False)
        print(f"\n💾 New file saved to: {output_path}")
        
        # Also save a version with only essential columns + EducationField
        print("\n📝 Creating version with EducationField...")
        
        # Add EducationField from raw data too
        current_df['EducationField'] = raw_df['EducationField']
        
        # Save complete version
        complete_path = '../data/cleaned/HR_PowerBI_Complete.csv'
        current_df.to_csv(complete_path, index=False)
        print(f"✅ Complete dataset with EducationField saved to: {complete_path}")
        
    else:
        print("\n❌ Data order doesn't match! Need a different approach.")
else:
    print(f"\n❌ Row counts don't match! Current: {len(current_df)}, Raw: {len(raw_df)}")

print("\n" + "=" * 50)
```
